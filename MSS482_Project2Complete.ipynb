> ## T-TEST:

```python
from scipy import stats

# Extract 'Bedroom' values for the '1' and '2' groups, excluding NaN
group_N = df_houses['price'][df_houses['Bedroom'] == 1].dropna()
group_Y = df_houses['price'][df_houses['Bedroom'] == 2].dropna()

# Perform t-test
statistic, p_value = stats.ttest_ind(group_N, group_Y, equal_var=False)

print(f"t's test statistic: {statistic}")
print(f"P-value: {p_value}")

alpha = 0.05

if p_value > alpha:
    print("The means of the two groups are approximately equal.")
else:
    print("The means of the two groups are significantly different.")
```

```python
from scipy import stats

# Extract 'Bathroom' values for the '1' and '2' groups, excluding NaN
group_N = df_houses['price'][df_houses['Bathroom'] == 1].dropna()
group_Y = df_houses['price'][df_houses['Bathroom'] == 2].dropna()

# Perform t-test
statistic, p_value = stats.ttest_ind(group_N, group_Y, equal_var=False)

print(f"t's test statistic: {statistic}")
print(f"P-value: {p_value}")

alpha = 0.05

if p_value > alpha:
    print("The means of the two groups are approximately equal.")
else:
    print("The means of the two groups are significantly different.")
```

```python
from scipy import stats

# Extract 'property size' values for the '1' and '2' groups, excluding NaN
group_N = df_houses['price'][df_houses['Property Size'] == 500].dropna()
group_Y = df_houses['price'][df_houses['Property Size'] == 1000].dropna()

# Perform t-test
statistic, p_value = stats.ttest_ind(group_N, group_Y, equal_var=False)

print(f"t's test statistic: {statistic}")
print(f"P-value: {p_value}")

alpha = 0.05

if p_value > alpha:
    print("The means of the two groups are approximately equal.")
else:
    print("The means of the two groups are significantly different.")
```

> ## LEVENE TEST

```python
median_size = df_houses['Property Size'].median()

# Create a new column 'Size Category' to categorize property size
df_houses['Size Category'] = df_houses['Property Size'].apply(lambda x: 'small' if x <= median_size else 'large')

# Now df_houses has a new column 'Size Category' with 'small' or 'large' labels
```

```python
from scipy.stats import levene

# Extract 'price' values for different groups
group_combinations = [('Bedroom', 1), ('Bedroom', 2), ('Bathroom', 1), ('Bathroom', 2), ('Size Category', 'small'), ('Size Category', 'large')]

alpha = 0.05

for group_var, group_val in group_combinations:
    # Extract 'price' values for the specific group
    group = df_houses['price'][df_houses[group_var] == group_val].dropna()
    # Extract 'price' values for all other groups
    other_groups = df_houses['price'][df_houses[group_var] != group_val].dropna()
    
    # Perform Levene test
    statistic, p_value = levene(group, other_groups)

    print(f"Levene test for {group_var}={group_val} - Statistic: {statistic}, P-value: {p_value}")

    if p_value > alpha:
        print("The variances are approximately equal.")
    else:
        print("The variances are significantly different.")
```

> ## SHAPIRO-WILK TEST

```python
import scipy.stats as stats
# List of variables to test
variables_to_test = ['Bedroom', 'Bathroom', 'Property Size']

for variable in variables_to_test:
    # Shapiro-Wilk test for each variable
    stat, p_value = stats.shapiro(df_houses[variable].dropna())

    print(f"Shapiro-Wilk test for '{variable}': Statistic={stat}, p-value={p_value}")

    alpha = 0.05  # Significance level

    if p_value > alpha:
        print(f"The '{variable}' variable looks like it came from a normal distribution (fail to reject H0)\n")
    else:
        print(f"The '{variable}' variable does not look like it came from a normal distribution (reject H0)\n")
```

### 1. Model Fit:

#### R-squared:
- **Analysis:** We computed the R-squared for the linear regression model. It measures how effectively the independent factors (bedroom, bathroom, and property size) explain the variability in the dependent variable (price).
- **Our Result:** Evaluate the R-squared value. A high R-squared value suggests a strong fit. If the value is near to one, it indicates that the chosen independent variables account for a considerable percentage of the variance in the dependent variable.

#### F-statistic and Prob(F-statistic):
- **Analysis:** We used the F-statistic and its associated p-value to determine the model's overall significance.
- **Our Result:** Check the p-value for the F-statistic. A low p-value indicates that at least one independent variable is important in predicting the dependent variable. This is critical to assessing the overall model significance.

### 2. Residuals:

#### Omnibus, Skewness, Kurtosis:
- **Analysis:** We examined various statistics (Omnibus, Skewness, Kurtosis) to assess the normality of the residuals.
- **Our Result:** Look for normality in residuals. A low Omnibus value suggests normal distribution. Additionally, skewness and kurtosis values close to 0 indicate that the residuals are symmetrically distributed.

#### Durbin-Watson:
- **Analysis:** We calculated the Durbin-Watson statistic to test for autocorrelation in the residuals.
- **Our Result:** A Durbin-Watson value around 2 suggests minimal autocorrelation. This is important for ensuring that the residuals are independent.

In summary, our analysis covers both model fit and residual diagnostics. A high R-squared and a low p-value for the F-statistic suggest a well-fitting model. Checking residuals for normality (low Omnibus, skewness, and kurtosis close to 0) and autocorrelation (Durbin-Watson around 2) is crucial for the reliability of the linear regression analysis. ### 8. T-test for Bedroom
The result of an independent t-test (also known as a two-sample t-test) conducted on the prices of houses with 1 bedroom versus houses with 2 bedrooms. This test is used to determine if there is a statistically significant difference between the means of two independent groups.

- **t's test statistic: -1.7507390334112738**: This is the calculated t-statistic for the test. The t-statistic measures the size of the difference relative to the variation in your sample data. The negative sign indicates that the mean of the first group (houses with 1 bedroom) is less than the mean of the second group (houses with 2 bedrooms). 

- **P-value: 0.08066140314625064**: The p-value is a measure of the probability that the observed difference between the group means would occur if the null hypothesis were true. The null hypothesis in this case states that there is no difference in the average price of houses with 1 bedroom and houses with 2 bedrooms. A p-value of 0.08066 is more than the commonly used significance level alpha of 0.05.

- **The means of the two groups are approximately equal.**: This conclusion is based on the p-value being greater than the significance level alpha (0.05). It implies that there is not enough statistical evidence to reject the null hypothesis. In other words, the difference in average prices between the two groups is not statistically significant at the 5% significance level.

It is crucial to note that "approximately equal" does not imply that the group means are identical or even near in value; rather, it indicates that any observed difference could be attributable to random chance rather than a systematic difference between the groups. . 

The decision to not reject the null hypothesis is a cautious one, indicating that we would need stronger evidence (a lower p-value) to confidently assert a real difference in the population means based on our sample data.
